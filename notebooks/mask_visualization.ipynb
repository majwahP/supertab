### Mask overlay visualization

This notebook visalises masks by overlaying them with the original data to see corresspondance

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
DS_FACTOR = 6
```

```python
# base = Image.open(f"../patch_outputs/patch_grid_ds{DS_FACTOR}_3d.png").convert("L")
base = Image.open(f"../stat_outputs/mask_grid_3d_ds{DS_FACTOR}_BMD_test4.png").convert("L")
mask = Image.open(f"../stat_outputs/mask_grid_3d_ds{DS_FACTOR}_BMD_test5.png").convert("L")

base_np = np.array(base)
mask_np = np.array(mask)

mask_norm = mask_np / 255.0

overlay = np.zeros((mask_np.shape[0], mask_np.shape[1], 4), dtype=np.float32)
overlay[..., 0] = 1.0 
overlay[..., 3] = 0.5 * mask_norm 

plt.figure(figsize=(8, 8), dpi=300)
plt.imshow(base_np, cmap='gray')
plt.imshow(overlay)
plt.axis('off')
plt.title("Overlay of Mask on Image")
plt.show()
```
